```python
final_dataset = df[['year', 'selling_price', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner']]
```

```python
final_dataset.head()
```

```python
# Deriving the age of car
final_dataset['current_year'] = 2021
```

```python
final_dataset['age_of_car'] = final_dataset['current_year']-final_dataset['year']
```

```python
final_dataset.head()
```

```python
final_dataset.drop(['year','current_year'],axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
final_dataset = pd.get_dummies(final_dataset,drop_first=True)
```

```python
final_dataset.head()
```

```python
final_dataset.corr()
```

```python
# Heatmap
corrmat = final_dataset.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
g = sns.heatmap(final_dataset[top_corr_features].corr(),annot=True,cmap='RdYlGn')
```

```python
X = final_dataset.iloc[:,1:]
y = final_dataset.iloc[:,0]
```

```python
X.head()
```

```python
X.columns
```

```python
# What are the most important features?
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
# Ploting graph of feature importance for better visualiation
feat_imp = pd.Series(model.feature_importances_,index=X.columns)
feat_imp.nlargest(10).plot(kind='barh') # Top 10
plt.show()
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Hyper parameters
# no of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=100,stop=1200,num=12)]
# no of features to consider at every split
max_features = ['auto','sqrt']
# maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(start=5,stop=30,num=6)]
# minimum number of samples required to split a node
min_samples_split = [2,5,10,15,1000]
# minimum number of samples required at each leaf node
min_samples_leaf = [1,2,5,10]
```

```python
random_grid = {
    'n_estimators' : n_estimators,
    'max_features' : max_features,
    'max_depth' : max_depth,
    'min_samples_split' : min_samples_split,
    'min_samples_leaf' : min_samples_leaf
}
print(random_grid)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Simple model because of size issue
rf.fit(X_train,y_train)
```

```python
rf_random = RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=10,
                              cv=5,verbose=2,random_state=42,n_jobs=1)
rf_random.fit(X_train,y_train)
```

```python
predictions2 = rf.predict(X_test)
```

```python
predictions = rf_random.predict(X_test)
```
